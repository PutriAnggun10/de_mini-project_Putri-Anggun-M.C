```python
import pandas as pd
import sqlite3
```

```python
# Memuat data CSV
gdp_data = pd.read_csv('D:\MSIB Anggun\mini-project\data_source\gdp_data.csv')
population_data = pd.read_csv('D:\MSIB Anggun\mini-project\data_source\population_data.csv')
electricity_access_percent = pd.read_csv('D:\MSIB Anggun\mini-project\data_source\electricity_access_percent.csv')
rural_population_percent = pd.read_csv('D:\MSIB Anggun\mini-project\data_source\rural_population_percent.csv')
projects_data = pd.read_csv('D:\MSIB Anggun\mini-project\data_source\projects_data.csv')

# Memuat data JSON
population_json = pd.read_json('D:\MSIB Anggun\mini-project\data_source\population_data.json')

# Cetak informasi singkat dari masing-masing DataFrame
print("GDP Data Summary:")
print(gdp_data.info())
print("Population Data Summary:")
print(population_data.info())
print("Electricity Access Percent Summary:")
print(electricity_access_percent.info())
print("Rural Population Percent Summary:")
print(rural_population_percent.info())
print("Projects Data Summary:")
print(projects_data.info())

print("Population JSON Data Summary:")
print(population_json.info())
```

```python
# Membuat koneksi ke database SQLite
conn = sqlite3.connect('D:\MSIB Anggun\mini-project\data_source\population_data.db')
cursor = conn.cursor()

# Menampilkan daftar semua tabel dalam database
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
print("Tables in the database:", tables)

# Meninjau struktur tabel pertama (misalnya)
table_name = tables[0][0]  # Asumsi tabel pertama
data_sample = pd.read_sql_query(f"SELECT * FROM {table_name} LIMIT 5", conn)
print(f"Data from {table_name}:")
print(data_sample.head())

# Tutup koneksi
conn.close()
```
